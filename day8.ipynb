## The scatter flot


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='xG', y='Goals', data=football_df)
plt.title('Goals vs Expected Goals (xG)')
plt.xlabel('Expected Goals (xG)')
plt.ylabel('Goals')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mins', y='Goals', data=football_df)
plt.title('Goals vs Minutes Played')
plt.xlabel('Minutes Played')
plt.ylabel('Goals')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Shots', y='Goals', data=football_df)
plt.title('Goals vs Shots')
plt.xlabel('Shots')
plt.ylabel('Goals')
plt.show()
```

## Population over the years

```python
population_df=pd.read_csv("world_population.csv")
asian_countries = population_df.loc[population_df["Continent"]=="Asia"].sort_values(by=["2022 Population"], ascending=False, ignore_index=True)
african_countries = population_df.loc[population_df["Continent"]=="Africa"].sort_values(by=["2022 Population"], ascending=False, ignore_index=True)
european_countries = population_df.loc[population_df["Continent"]=="Europe"].sort_values(by=["2022 Population"], ascending=False, ignore_index=True)
na_countries = population_df.loc[population_df["Continent"]=="North America"].sort_values(by=["2022 Population"], ascending=False, ignore_index=True)
oc_countries = population_df.loc[population_df["Continent"]=="Oceania"].sort_values(by=["2022 Population"], ascending=False, ignore_index=True)
sa_countries = population_df.loc[population_df["Continent"]=="South America"].sort_values(by=["2022 Population"], ascending=False, ignore_index=True)
continent_df = population_df.groupby(by='Continent').sum()
altered_continent_df = continent_df.drop(columns=["CCA3", "Country/Territory", "Rank", "Capital", "Growth Rate", "Area (km)", "Density (per km)","World Population Percentage"])
continent_df = population_df.groupby(by='Continent').sum()
altered_continent_df = continent_df.drop(columns=["CCA3", "Country/Territory", "Rank", "Capital", "Growth Rate", "Area (km)", "Density (per km)","World Population Percentage"])
transposed = altered_continent_df.T
transposed.index = transposed.index.str.replace(" Population", "")
transposed.sort_index().plot.line(title="Population over the years", legend=True)
plt.show()
```
